```python
import pandas as pd
import numpy as np
```

```python
# 10 dòng đầu
df = pd.read_csv('/content/sample_data/dulieuxettuyendaihoc.csv')
df.head(10)
```

```python
# 10 dòng cuối
df.tail(10)
```

```python
#Thay thế dữ liệu thiếu
df['DT'].fillna(0, inplace=True)
df
```

```python
# xử lý lần lượt tất cả dữ liệu thiếu cho các biến về điểm số còn lại
cac_bien_con_lai = [
    'L1', 'H1', 'S1', 'V1', 'X1', 'D1', 'N1',
    'T2', 'L2', 'H2', 'S2', 'V2', 'X2', 'D2', 'N2',
    'T6', 'L6', 'H6', 'S6', 'V6', 'X6', 'D6', 'N6',
    'DH1', 'DH2', 'DH3'
]
for col in cac_bien_con_lai:
    df[col].fillna(df[col].mean(), inplace=True)
    df
```

```python
#Tạo các biến TBM1, TBM2, TBM3 tương ứng với trung bình môn của các năm lớp 10, 11 và
12.
df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10
df['TBM2'] = (df['T2']*2 + df['L2'] + df['H2'] + df['S2'] + df['V2']*2 + df['X2'] + df['D2'] + df['N2']) / 10
df['TBM3'] = (df['T6']*2 + df['L6'] + df['H6'] + df['S6'] + df['V6']*2 + df['X6'] + df['D6'] + df['N6']) / 10

print(df[['TBM1', 'TBM2', 'TBM3']])
```

```python
#Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1,TBM2 và TBM3
def phan_loai_hoc_luc(diem):
    if diem >= 9:
        return 'Xuất Sắc'
    elif diem >= 8:
        return 'Giỏi'
    elif diem >= 7:
        return 'Khá'
    elif diem >= 5:
        return 'Trung Bình'
    else:
        return 'Yếu'

df["XL1"] = df["TBM1"].apply(phan_loai_hoc_luc)
df["XL2"] = df["TBM2"].apply(phan_loai_hoc_luc)
df["XL3"] = df["TBM3"].apply(phan_loai_hoc_luc)


df


```

```python
#Tạo các biến US_TBM1, US_TBM2 và US_TBM3
df['US_TBM1'] = (df['TBM1'] / 10) * 4
df['US_TBM2'] = (df['TBM2'] / 10) * 4
df['US_TBM3'] = (df['TBM3'] / 10) * 4
df
```

```python
#Tạo biến kết quả xét tuyển (kí hiệu là KQXT) nhằm xác định sinh viên đậu (giá trị “1”) và rớt ( giá trị “0”) vào các khối dựa trên điểm DH1, DH2 và DH3
def ket_qua_xet_tuyen(du_lieu):
  khoi_thi =du_lieu['KT']
  dh1 = du_lieu['DH1']
  dh2 = du_lieu['DH2']
  dh3 = du_lieu['DH3']



  if khoi_thi in ['A','A1']:
    diem_xt = (dh1 * 2 + dh2 + dh3)/4
  elif khoi_thi in ['B','B1']:
      diem_xt = (dh1 + dh2*2 + dh3)/4
  else:
      diem_xt = (dh1 + dh2 + dh3*2) / 3

  if diem_xt >= 5.0:
      return 1
  else:
      return 0

df['KQXT'] = df.apply(ket_qua_xet_tuyen, axis=1)
df
```
